```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python


def df_dt(x, t, a, b, c, d):
    """Función del sistema en forma canónica"""
    dx = a * x[0] - b * x[0] * x[1]
    dy = - c * x[1] + d * x[0] * x[1]
    return np.array([dx, dy])

# Parámetros
a = 0.1
b = 0.02
c = 0.3
d = 0.01

# Condiciones iniciales
x0 = 40   # Presas
y0 = 9    # Depredadores
conds_iniciales = np.array([x0, y0])

# Condiciones para integración
tf = 200
N = 800
t = np.linspace(0, tf, N)

solucion = odeint(df_dt, conds_iniciales, t, args=(a, b, c, d))
#plt.plot(t, solucion[:, 0], label='presa')
#plt.plot(t, solucion[:, 1], label='depredador')
plt.plot(solucion[:, 0], solucion[:, 1])
plt.xlabel('Presas', fontsize = 14) 
plt.ylabel('Depredadores', fontsize = 14) 
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def pend(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

b = 0.25
c = 5.0

y0 = [np.pi - 0.1, 0.0]

t = np.linspace(0, 10, 101)


sol = odeint(pend, y0, t, args=(b, c))


plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def sys1(x, t ):
    x1, x2 = x
    a = 1
    b = .5
    dx1dt = x1*(1 -x1 -a*x2)
    dx2dt = b*x2*(x1 -x2)
    return [dx1dt, dx2dt]


x0 = [1,0]#[1, 2]

t = np.arange(0,100,.1)


sol = odeint(sys1, x0, t)


plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
import phaseportrait

def sys1phase(x1,x2 ):
    #x1, x2 = x
    a = 1
    b = .5
    dx1dt = x1*(1 -x1 -a*x2)
    dx2dt = b*x2*(x1 -x2)
    return [dx1dt, dx2dt]

SimplePendulum = phaseportrait.PhasePortrait2D(sys1phase, [-9, 9], Title='Simple sys', xlabel=r"$x_1$", ylabel=r"$\dot{x_2}$")
SimplePendulum.plot()
```
